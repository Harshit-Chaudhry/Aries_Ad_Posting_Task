```python
import pyautogui
import time

# Configuration
# X=1210, Y=1009
X_POSITION = 1210  # X coordinate where to click
Y_POSITION = 1009  # Y coordinate where to click
TEXT_TO_TYPE = "post_invite_card clan_WA3K4A"  # Text to type
REPEAT_INTERVAL = 5  # Time in seconds between repetitions
NUM_REPETITIONS = 10  # Number of times to repeat (set to -1 for infinite)

# Safety settings
pyautogui.FAILSAFE = True  # Move mouse to corner to stop
pyautogui.PAUSE = 0.5  # Pause between actions

def perform_action():
    """Perform the mouse click, typing, and enter sequence."""
    # Move mouse to position
    pyautogui.moveTo(X_POSITION, Y_POSITION, duration=0.5)
    
    # Click at the position
    pyautogui.click()
    
    # Wait a moment for any UI to respond
    time.sleep(0.3)
    
    # Type the text
    pyautogui.typewrite(TEXT_TO_TYPE, interval=0.05)
    
    # Press Enter
    pyautogui.press('enter')
    
    print(f"Action completed: Clicked at ({X_POSITION}, {Y_POSITION}), typed '{TEXT_TO_TYPE}', and pressed Enter")

def main():
    print("Starting automation script...")
    print(f"Will repeat every {REPEAT_INTERVAL} seconds")
    print("Move mouse to top-left corner to emergency stop (FAILSAFE)")
    print("\nStarting in 3 seconds...")
    time.sleep(3)
    
    repetitions = 0
    try:
        while True:
            # Perform the action
            perform_action()
            
            repetitions += 1
            
            # Check if we've reached the desired number of repetitions
            if NUM_REPETITIONS != -1 and repetitions >= NUM_REPETITIONS:
                print(f"\nCompleted {repetitions} repetitions. Exiting.")
                break
            
            # Wait before next repetition
            print(f"Waiting {REPEAT_INTERVAL} seconds before next repetition...\n")
            time.sleep(REPEAT_INTERVAL)
            
    except KeyboardInterrupt:
        print("\n\nScript stopped by user (Ctrl+C)")
    except pyautogui.FailSafeException:
        print("\n\nFAILSAFE triggered! Mouse moved to corner. Script stopped.")

if __name__ == "__main__":
    # Display current mouse position helper
    print("=== Mouse Position Helper ===")
    print("Move your mouse to the desired position and note the coordinates below:")
    for i in range(5, 0, -1):
        x, y = pyautogui.position()
        print(f"Current position: X={x}, Y={y} ({i}s remaining)", end='\r')
        time.sleep(1)
    print("\n")
    
    main()
```
